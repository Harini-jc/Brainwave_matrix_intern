```python
# Inventory Management System

import tkinter as tk
from tkinter import messagebox
import json

class InventoryManager:
    def _init_(self):
        self.products = {}
        self.load_data()

    def load_data(self):
        try:
            with open('inventory.json', 'r') as file:
                self.products = json.load(file)
        except FileNotFoundError:
            self.products = {}

    def save_data(self):
        with open('inventory.json', 'w') as file:
            json.dump(self.products, file)

    def add_product(self, name, quantity, price):
        self.products[name] = {'quantity': quantity, 'price': price}
        self.save_data()

    def edit_product(self, name, quantity, price):
        if name in self.products:
            self.products[name] = {'quantity': quantity, 'price': price}
            self.save_data()
        else:
            raise ValueError("Product not found.")

    def delete_product(self, name):
        if name in self.products:
            del self.products[name]
            self.save_data()
        else:
            raise ValueError("Product not found.")

    def low_stock_alert(self, threshold):
        return {name: details for name, details in self.products.items() if details['quantity'] < threshold}

    def sales_summary(self):
        return {name: details['quantity'] * details['price'] for name, details in self.products.items()}

class InventoryApp:
    def _init_(self, root):
        self.manager = InventoryManager()
        self.root = root
        self.root.title("Inventory Management System")
        self.create_widgets()

    def create_widgets(self):
        self.name_label = tk.Label(self.root, text="Product Name")
        self.name_label.pack()
        self.name_entry = tk.Entry(self.root)
        self.name_entry.pack()

        self.quantity_label = tk.Label(self.root, text="Quantity")
        self.quantity_label.pack()
        self.quantity_entry = tk.Entry(self.root)
        self.quantity_entry.pack()

        self.price_label = tk.Label(self.root, text="Price")
        self.price_label.pack()
        self.price_entry = tk.Entry(self.root)
        self.price_entry.pack()

        self.add_button = tk.Button(self.root, text="Add Product", command=self.add_product)
        self.add_button.pack()

        self.edit_button = tk.Button(self.root, text="Edit Product", command=self.edit_product)
        self.edit_button.pack()

        self.delete_button = tk.Button(self.root, text="Delete Product", command=self.delete_product)
        self.delete_button.pack()

        self.report_button = tk.Button(self.root, text="Low Stock Alert", command=self.show_low_stock)
        self.report_button.pack()

        self.summary_button = tk.Button(self.root, text="Sales Summary", command=self.show_sales_summary)
        self.summary_button.pack()

    def add_product(self):
        name = self.name_entry.get()
        quantity = int(self.quantity_entry.get())
        price = float(self.price_entry.get())
        self.manager.add_product(name, quantity, price)
        messagebox.showinfo("Success", "Product added successfully.")

    def edit_product(self):
        name = self.name_entry.get()
        quantity = int(self.quantity_entry.get())
        price = float(self.price_entry.get())
        try:
            self.manager.edit_product(name, quantity, price)
            messagebox.showinfo("Success", "Product edited successfully.")
        except ValueError as e:
            messagebox.showerror("Error", str(e))

    def delete_product(self):
        name = self.name_entry.get()
        try:
            self.manager.delete_product(name)
            messagebox.showinfo("Success", "Product deleted successfully.")
        except ValueError as e:
            messagebox.showerror("Error", str(e))

    def show_low_stock(self):
        threshold = 5  # Example threshold
        low_stock = self.manager.low_stock_alert(threshold)
        messagebox.showinfo("Low Stock Alert", str(low_stock))

    def show_sales_summary(self):
        summary = self.manager.sales_summary()
        messagebox.showinfo("Sales Summary", str(summary))

if _name_ == "_main_":
    root = tk.Tk()
    app = InventoryApp(root)
    root.mainloop()
```
